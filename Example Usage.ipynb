# ElexonDataPortal Example Usage

Thie notebook walks through a number of examples of how to use the ElexonDataPortal

The ElexonDataPortal provides the ability to download data directly from the Elexon API and have it parsed into a user-friendly Pandas dataframe. All queries require both a stream name and query arguments, all query argument information can be found <a href="https://www.elexon.co.uk/wp-content/uploads/2018/09/BMRS-API-Data-Push-User-Guide.pdf">here</a>. Previous users of Elexon data will have had to switch between many different naming conventions for query arguments, the ElexonDataPortal harmonises these so that instead of using one of say <i>FromDate</i>, <i>EventStart</i> or <i>SettlementDate</i> we would use <i>start_date</i> for all of them.

<br>

### Imports

````python
## Data Manipulation
import pandas as pd
import numpy as np

## Plotting
import seaborn as sns
import matplotlib.pyplot as plt

## Custom 
import ElexonDataPortal.API as edp_API
from ElexonDataPortal import cleaner, plotter
````

<br>

### Initialising API Wrapper

Before you can use the ElexonDataPortal module you'll have to set-up an API key with Elexon which can be done <a href="https://www.elexonportal.co.uk/registration/newuser">here</a>

````python
API_key = 'your_password_here'

edp_wrapper = edp_API.Wrapper(API_key)
````

<br>

### Making Queries

Elexon allows you to query data in a very limited fashion, often constrained to returning individual settlment periods. These queries are exposed through the default <i>query</i> function from the API wrapper.

Each query requires a data stream name as well as more specific arguments which are listed in the <a href="https://www.elexon.co.uk/wp-content/uploads/2018/09/BMRS-API-Data-Push-User-Guide.pdf">API guide</a>.

````python
stream = 'B1630'

query_args = {
    'query_date' : '2019-09-25',
    'SP' : 48,
}
    
df_RES_SP = edp_wrapper.query(stream, query_args)

df_RES_SP
````

<br>

Often though we're more interested in getting data for a whole range of settlement periods, this is easy to achieve through the query orchestrator which accepts a start and end date alongside any other query specific arguments.

````python
stream = 'B1630'

query_args = {
    'start_date' : '2019-09-25',
    'end_date' : '2019-09-25 23:30',
}

df_RES = edp_wrapper.query_orchestrator(stream, query_args)

df_RES.head(3)
````

<br>

Here we'll parse the data from its raw format

````python
clean_column_names = lambda df: df.rename(columns=dict(zip(df.columns, (df.columns
                                                                        .str.replace('"', '')
                                                                        .str.replace(' ', '_')
                                                                        .str.lower()))))

df_RES_cleaned = (df_RES
                  .pivot('datetime', 'powerSystemResourceType', 'quantity') 
                  .pipe(clean_column_names)
                  .astype(float)                 
                 )

df_RES_cleaned.head()
````

<br>

Note that the <i>B1630</i> data stream must be used with caution as it often reports incorrect wind generation figures (shown in the spiky output), these errors have propogated to other sources such as ENTSOE or electricinsights. It is recommended that you use either the aggregated wind generation from <i>FUELHH</i> or sum up the individual on- and off-shore wind farms generation.

````python
df_RES_cleaned.plot()
````

<br>

### More Complex Arguments

We can use the same querying format to retrieve data for individual unit generation. Note that we had to provide an additionaly argument to the query, <i>BMU_id</i>, which specifies the BMU generator we wish to retrieve information for. If you want to query for all BMUs then use ```'*'``` in place of an id string. Again all query argument information can be found <a href="https://www.elexon.co.uk/wp-content/uploads/2018/09/BMRS-API-Data-Push-User-Guide.pdf">here</a>.

````python
stream = 'B1610'

query_args = {
    'start_date' : '2019-09-25',
    'end_date' : '2019-09-25 23:30',
    'BMU_id' : '*'
}

df_unit_gen = edp_wrapper.query_orchestrator(stream, query_args)

df_unit_gen.head(3)
````

<br>

We could instead provide a specific BMU_id, in this example we'll download data for 1/4 of the London Array wind farm

````python
stream = 'B1610'

query_args = {
    'start_date' : '2019-09-25',
    'end_date' : '2019-09-25 23:30',
    'BMU_id' : 'LARYO-1'
}

df_LARYO1_gen = edp_wrapper.query_orchestrator(stream, query_args)

df_LARYO1_gen.head(3)
````

<br>

We'll quickly clean this up and visualise it

````python
df_LARYO1_gen_cleaned = (df_LARYO1_gen
                         [~df_LARYO1_gen['settlementPeriod'].isnull()] # Filtering for rows specific to a SP
                         .set_index('datetime') 
                         ['quantity']
                         .astype(float)
                        )

df_LARYO1_gen_cleaned.plot()
````

### Using the ElexonDataPortal Helpers

The API wrapper has been designed so that you always receive the full set of data that Elexon returns, if you want to help contribute to this work one area that is highly beneficial is developing helper functions for common data extraction use cases.

One example stream with pre-built cleaning and plotting functions is DETSYSPRICES which contains all of the detailed system prices. We'll start by downloading data for just a single settlement period.

````python
stream = 'DETSYSPRICES'

query_args = {
    'query_date' : '2015-01-01', 
    'SP' : '1',
}

df_accepted_BOs = edp_wrapper.query(stream, query_args)

df_accepted_BOs.head()
````

````python
stream = 'FUELHH'

query_args = {
    'start_date' : '2019-09-25',
    'end_date' : '2019-09-25 23:30',
}

df_generation = edp_wrapper.query_orchestrator(stream, query_args)

df_generation.head()

df_generation.head().to_html().replace('\n', '')
````

<br>

We'll now use a helper function to convert the Bid Offer Data (BOD) into a price volume stack

````python
df_imb_stack = cleaner.accepted_BODs_2_stack(df_accepted_BOs[df_accepted_BOs['soFlag']=='F'])

df_imb_stack.head()
````

<br>

Which can also be easily visualised

````python
plotter.plot_BOD_stack(df_imb_stack)
````

<br>

There are further helpers such as the <i>dt_rng_2_SPs</i> function which creates a mapping from a Pandas Timestamp to any date and settlement period

````python
df_date_2_SP = edp_wrapper.dt_rng_2_SPs(start_date='2019-09-01', end_date='2019-09-01 23:30')

df_date_2_SP.head()
````
